# Causal Inference in Time Series Data

Prediction of risk from annual report is a tricky problem to solve and we aim to use causal graph modeling to do the prediction.


Assumption:

1. Causal graph can be established from demain knowledge.
2. The previous years' report resulst can have a direct effect on the future years.
3. The relationship formulates exatly a DAGs and there is no hidden variables other than variables targeted.

Methods:

1. Causal graph discovery.
2. Causal modeling.
3. Causal estimation for the next year.

Data:

1.Data are from public available reports and then extracted using an open-source tools.
2.Graph represents the direct relation from previous year to next year.

## Graph Representation

```python
import dowhy
from dowhy import CausalModel

import numpy as np
import pandas as pd
import graphviz
import networkx as nx
from causallearn.search.FCMBased.lingam.utils import make_dot

np.set_printoptions(precision=3, suppress=True)
np.random.seed(0)

import numpy as np
from causallearn.search.FCMBased import lingam
from causallearn.search.FCMBased.lingam.utils import make_dot

labels = ["D1", "D2", "D3", "D4", "E1", "E2", "E3", "E4", "F1", "F2", "F3", "F4"]

size = len(labels)
adjacency_matrix = np.zeros((size, size), dtype=int)

connections = ["D3", "D2", "D4", "E2", "E3", "E4", "F2", "F3", "F4"]
for node in ["D1"]:
    for connection in connections:
        adjacency_matrix[labels.index(connection)][labels.index(node)] = 1


connections = ["D3", "E3", "E4", "F3", "F4", "D4"]
for node in ["D2"]:
    for connection in connections:
        adjacency_matrix[labels.index(connection)][labels.index(node)] = 1

connections = ["D4", "E4", "F4"]
for node in ["D3"]:
    for connection in connections:
        adjacency_matrix[labels.index(connection)][labels.index(node)] = 1


make_dot(adjacency_matrix, labels=labels)
```
